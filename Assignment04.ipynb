```python
n = 1
sentence = 'Three cats chased three playful mice around the garden. Five colorful balloons floated high above the carnival. Seven students eagerly awaited their turn to present. Two tall trees swayed gently in the evening breeze'
unigrams = ngrams(sentence.split(), n)

for item in unigrams:
    print(item)
```

```python
from nltk import ngrams
file = open("text1.txt")
for i in file.readlines():
    cumulative = i
    sentences = i.split(".")
    counter = 0
    for sentence in sentences:
        print("For sentence", counter + 1, ", trigrams are: ")
        trigrams = ngrams(sentence.split(" "), 3)
        for grams in trigrams:
            print(grams)
        counter += 1
        print()
```
